```python
import pandas as pd
```

```python
zombies_table = pd.read_html('https://en.wikipedia.org/wiki/List_of_zombie_films')
zombies_df = pd.DataFrame(zombies_table[0])
zombies_df['Monster'] = "Zombie"
del zombies_df['Notes']
del zombies_df['References']
del zombies_df['Director']

zombies_df.head()
```

```python
ghost_table = pd.read_html('https://en.wikipedia.org/wiki/List_of_ghost_films')
ghost_df = pd.DataFrame(ghost_table[1]).rename(columns={"Name":"Title"})
del ghost_df['Country, notes']
ghost_df['Monster'] = "Ghost"
ghost_df.head()
```

```python
vampire_table = pd.read_html('https://en.wikipedia.org/wiki/List_of_vampire_films')
vampire_df = pd.DataFrame(vampire_table[1]).rename(columns={"Release date":"Year"})
del vampire_df ['Director']
del vampire_df ['Leading cast']
del vampire_df ['Country']
del vampire_df ['Notes']
vampire_df['Monster'] = "Vampire"
vampire_df.head()
```

```python
monster_table = pd.read_html('https://en.wikipedia.org/wiki/List_of_monster_movies')
monster_df = pd.DataFrame(monster_table[0]).rename(columns={"Name":"Title","Release year":"Year", "Monster(s)": "Monster"})
del monster_df ['Ref.']
monster_df.head()
```
